```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
```

```python
genres = ["Drama", "Comedy", "Action", "Romance", "Thriller", "Musical"]

positive_reviews = [
    "An outstanding film with great performances.",
    "Loved the storyline and direction!",
    "A must-watch for movie lovers.",
    "Brilliantly executed and very entertaining."
]

neutral_reviews = [
    "It was okay, not too bad.",
    "Average movie with some good moments.",
    "Decent watch but nothing special.",
    "Mediocre plot but watchable."
]

negative_reviews = [
    "Disappointing and poorly executed.",
    "Weak storyline and dull acting.",
    "Not worth the time.",
    "Failed to keep me engaged."
]
```

```python
data = []
for i in range(1000):
    title = fake.catch_phrase()
    year = random.randint(2000, 2025)
    genre = random.choice(genres)
    director = fake.name()
    lead = fake.name()
    rating = round(random.uniform(1.0, 10.0), 1)

    recommendation = "Yes" if rating >= 4.0 else "No"

    if rating >= 7.0:
        review = random.choice(positive_reviews)
    elif rating >= 4.0:
        review = random.choice(neutral_reviews)
    else:
        review = random.choice(negative_reviews)

    data.append([i+1, title, year, genre, director, lead, rating, recommendation, review])
```

```python
df = pd.DataFrame(data, columns=[
    "Movie ID", "Title", "Release Year", "Genre", 
    "Director", "Lead", "IMDB Rating", "Recommendation", "Review"
])
```

```python
df.set_index("Movie ID", inplace=True)
```

```python
df.to_csv("fictional_movies.csv", index=False)
```
